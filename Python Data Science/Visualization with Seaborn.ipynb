# Seaborn vs. Matplotlib

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
#random walk data
rng = np.random.RandomState(0)
x = np.linspace(0,10,500)
y = np.cumsum(rng.randn(500,6),0)
```

```python
#plot the data with matplotlib defaults:old-fashioned
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left');
```

```python
#plot the data with seaborn:much better
import seaborn as sns
sns.set()

plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left');
```

```python
#exploring seaborn plots
#1. histogram
data = np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data = pd.DataFrame(data,columns=['x','y'])

for col in 'xy':
    plt.hist(data[col],alpha=0.5);
```

```python
#2. smooth estimate of the distribution using a kernel density estimation
for col in 'xy':
    sns.kdeplot(data[col], shade=True);
```

```python
#3. combine histogarm and density
for col in 'xy':
    sns.distplot(data[col]);
```

```python
#kde plot of data
sns.kdeplot(data);
```

```python
#joint dist with marginal dist
with sns.axes_style('white'):
    sns.jointplot('x','y',data,kind='kde');
```

```python
#pairwise plots with iris data
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris,hue='species',size=2.5);
```

```python
#Faceted histograms with tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
#use sns,FacetGrid
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

```python
#Factor plots
with sns.axes_style(style='ticks'):
    g = sns.factorplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total Bill");
```

```python
#joint distributions
sns.jointplot('total_bill','tip',data=tips,kind='reg');
```

```python
#barplot: time series
planets = sns.load_dataset('planets')
planets.head()
```

```python
#factor plot
g = sns.factorplot('year',data=planets, aspect=2,kind='count',color='steelblue')
g.set_xticklabels(step=5);
```

```python
#hue='method',to look at the method of discovery of each of these planets
g = sns.factorplot("year", data=planets, aspect=4.0, kind='count',
                       hue='method', order=range(2001, 2015))
g.set_ylabels('Number of Planets Discovered');
```

```python
#violinplot
sns.violinplot('method','mass',data=planets,inner='quartile',palette=['lightblue','lightpink']);
```
